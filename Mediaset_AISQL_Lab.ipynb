# Laboratorio AI SQL - Mediaset

Benvenuti nel laboratorio pratico sulle **funzioni AI SQL di Snowflake Cortex** applicate ai dati del broadcasting televisivo Mediaset.

In questo notebook esploreremo **6 funzioni AI SQL** che permettono di analizzare testi, generare contenuti e classificare informazioni direttamente in SQL:

1. **AI_SENTIMENT** - Analisi del sentiment dei feedback social
2. **AI_SUMMARIZE** - Riassunto automatico delle descrizioni dei contenuti
3. **AI_TRANSLATE** - Traduzione multilingua dei testi
4. **AI_COMPLETE** - Generazione di testo e analisi avanzate con LLM
5. **AI_CLASSIFY_TEXT** - Classificazione automatica dei feedback
6. **AI_EXTRACT_ANSWER** - Estrazione di informazioni specifiche dai testi

## 1. Sentiment Analysis

La funzione **AI_SENTIMENT** analizza il tono emotivo di un testo e restituisce un punteggio numerico compreso tra -1 (molto negativo) e +1 (molto positivo).

Applicheremo questa funzione ai feedback social dei telespettatori per capire come vengono percepiti i programmi Mediaset sulle diverse piattaforme.

## 2. Summarize

La funzione **AI_SUMMARIZE** genera automaticamente un riassunto conciso di un testo lungo.

Utilizzeremo questa funzione per creare riassunti delle descrizioni complete dei contenuti televisivi, utili per schede rapide o anteprime.

## 3. Translate

La funzione **AI_TRANSLATE** traduce testi da una lingua all'altra supportando numerose combinazioni linguistiche.

Tradurremo i feedback social e le descrizioni dei contenuti in inglese, spagnolo e francese per supportare la distribuzione internazionale.

## 4. AI Complete

La funzione **AI_COMPLETE** utilizza modelli LLM (Large Language Models) per generare testo, analizzare dati e produrre insight a partire da un prompt.

Utilizzeremo il modello **mistral-large2** per generare analisi editoriali sui feedback e mini-report sui dati di ascolto.

## 5. Classify Text

La funzione **AI_CLASSIFY_TEXT** classifica automaticamente un testo in una delle categorie fornite.

Classificheremo i feedback dei telespettatori in categorie editoriali predefinite per facilitare il triage e la gestione delle segnalazioni.

## 6. Extract Answer

La funzione **AI_EXTRACT_ANSWER** estrae risposte specifiche da un testo in base a una domanda posta in linguaggio naturale.

Utilizzeremo questa funzione per estrarre informazioni strutturate dai feedback e dalle descrizioni dei contenuti.

### Punti chiave
- Tutte le funzioni operano **direttamente in SQL**, senza necessita' di infrastruttura ML esterna

- Le funzioni sono **combinabili** tra loro per creare pipeline di analisi complesse
- L'integrazione con i dati strutturati (JOIN, aggregazioni) permette analisi a **360 gradi**
